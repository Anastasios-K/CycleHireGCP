```python
import yaml
import pandas as pd
from IPython.display import display
```

```python
class YamlReader:

    @staticmethod
    def read_yaml_file(path: str):
        """ Read yaml file. """
        with open(path) as file:
            content = yaml.load(file, Loader=yaml.FullLoader)
            file.close()
        return content
```

```python
class InfoTracker:
    """ 
    Module used throughout the pipeline to track useful information.
    A number of info is used in later stages of the pipeline.
    """
    def __init__(self,
                 cycle_hire_preview: pd.DataFrame = None,
                 cycle_stations_preview: pd.DataFrame = None,
                 hires_null_values_count: pd.DataFrame = None,
                 stations_null_values_count: pd.DataFrame = None,
                 total_rides_n_duration_per_year: pd.DataFrame = None,
                 busiest_starting_stations_in_rides: pd.DataFrame = None,
                 least_busy_starting_stations_in_rides: pd.DataFrame = None,
                 busiest_starting_stations_in_rides_per_year: pd.DataFrame = None,
                 most_profitable_starting_station_per_year: pd.DataFrame = None,
                 top_destinations_of_all_time: pd.DataFrame = None,
                 top_destinations_per_year: pd.DataFrame = None,
                 most_popular_roots_of_all_time: pd.DataFrame = None,
                 top_roots_per_year: pd.DataFrame = None,
                 daily_n_weekly_usage_pattern: pd.DataFrame = None,
                 cycle_station_data_with_borough_names: pd.DataFrame = None,
                 cycle_station_with_borough_names_preview: pd.DataFrame = None,
                 total_duartion_per_borough: pd.DataFrame = None,
                 h2o_leaderboard = None,
                 h2o_leaderboard_df: pd.DataFrame = None
                 ):

        self.__cycle_hire_preview = cycle_hire_preview
        self.__cycle_stations_preview = cycle_stations_preview
        self.__hires_null_values_count = hires_null_values_count
        self.__stations_null_values_count = stations_null_values_count
        self.__total_rides_n_duration_per_year = total_rides_n_duration_per_year
        self.__busiest_starting_stations_in_rides = busiest_starting_stations_in_rides
        self.__least_busy_starting_stations_in_rides = least_busy_starting_stations_in_rides
        self.__busiest_starting_stations_in_rides_per_year = busiest_starting_stations_in_rides_per_year
        self.__most_profitable_starting_station_per_year = most_profitable_starting_station_per_year
        self.__top_destinations_of_all_time = top_destinations_of_all_time
        self.__top_destinations_per_year = top_destinations_per_year
        self.__most_popular_roots_of_all_time = most_popular_roots_of_all_time
        self.__top_roots_per_year = top_roots_per_year
        self.__daily_n_weekly_usage_pattern = daily_n_weekly_usage_pattern
        self.__cycle_station_data_with_borough_names = cycle_station_data_with_borough_names
        self.__cycle_station_with_borough_names_preview = cycle_station_with_borough_names_preview
        self.__total_duartion_per_borough = total_duartion_per_borough
        self.__h2o_leaderboard = h2o_leaderboard
        self.__h2o_leaderboard_df = h2o_leaderboard_df
    
    @property
    def cycle_hire_preview(self):
        return self.__cycle_hire_preview

    @cycle_hire_preview.setter
    def cycle_hire_preview(self, value: pd.DataFrame):
        self.__cycle_hire_preview = value

    @property
    def cycle_stations_preview(self):
        return self.__cycle_stations_preview

    @cycle_stations_preview.setter
    def cycle_stations_preview(self, value: pd.DataFrame):
        self.__cycle_stations_preview = value

    @property
    def hires_null_values_count(self):
        return self.__hires_null_values_count

    @hires_null_values_count.setter
    def hires_null_values_count(self, value: pd.DataFrame):
        self.__hires_null_values_count = value

    @property
    def stations_null_values_count(self):
        return self.__stations_null_values_count

    @stations_null_values_count.setter
    def stations_null_values_count(self, value: pd.DataFrame):
        self.__stations_null_values_count = value

    @property
    def total_rides_n_duration_per_year(self):
        return self.__total_rides_n_duration_per_year

    @total_rides_n_duration_per_year.setter
    def total_rides_n_duration_per_year(self, value: pd.DataFrame):
        self.__total_rides_n_duration_per_year = value

    @property
    def busiest_starting_stations_in_rides(self):
        return self.__busiest_starting_stations_in_rides

    @busiest_starting_stations_in_rides.setter
    def busiest_starting_stations_in_rides(self, value: pd.DataFrame):
        self.__busiest_starting_stations_in_rides = value

    @property
    def least_busy_starting_stations_in_rides(self):
        return self.__least_busy_starting_stations_in_rides

    @least_busy_starting_stations_in_rides.setter
    def least_busy_starting_stations_in_rides(self, value: pd.DataFrame):
        self.__least_busy_starting_stations_in_rides = value

    @property
    def busiest_starting_stations_in_rides_per_year(self):
        return self.__busiest_starting_stations_in_rides_per_year

    @busiest_starting_stations_in_rides_per_year.setter
    def busiest_starting_stations_in_rides_per_year(self, value: pd.DataFrame):
        self.__busiest_starting_stations_in_rides_per_year = value

    @property
    def most_profitable_starting_station_per_year(self):
        return self.__most_profitable_starting_station_per_year

    @most_profitable_starting_station_per_year.setter
    def most_profitable_starting_station_per_year(self, value: pd.DataFrame):
        self.__most_profitable_starting_station_per_year = value

    @property
    def top_destinations_of_all_time(self):
        return self.__top_destinations_of_all_time

    @top_destinations_of_all_time.setter
    def top_destinations_of_all_time(self, value: pd.DataFrame):
        self.__top_destinations_of_all_time = value

    @property
    def top_destinations_per_year(self):
        return self.__top_destinations_per_year

    @top_destinations_per_year.setter
    def top_destinations_per_year(self, value: pd.DataFrame):
        self.__top_destinations_per_year = value

    @property
    def most_popular_roots_of_all_time(self):
        return self.__most_popular_roots_of_all_time

    @most_popular_roots_of_all_time.setter
    def most_popular_roots_of_all_time(self, value: pd.DataFrame):
        self.__most_popular_roots_of_all_time = value

    @property
    def top_roots_per_year(self):
        return self.__top_roots_per_year

    @top_roots_per_year.setter
    def top_roots_per_year(self, value: pd.DataFrame):
        self.__top_roots_per_year = value

    @property
    def daily_n_weekly_usage_pattern(self):
        return self.__daily_n_weekly_usage_pattern

    @daily_n_weekly_usage_pattern.setter
    def daily_n_weekly_usage_pattern(self, value: pd.DataFrame):
        self.__daily_n_weekly_usage_pattern = value

    @property
    def cycle_station_data_with_borough_names(self):
        return self.__cycle_station_data_with_borough_names

    @cycle_station_data_with_borough_names.setter
    def cycle_station_data_with_borough_names(self, value: pd.DataFrame):
        self.__cycle_station_data_with_borough_names = value

    @property
    def cycle_station_with_borough_names_preview(self):
        return self.__cycle_station_with_borough_names_preview

    @cycle_station_with_borough_names_preview.setter
    def cycle_station_with_borough_names_preview(self, value: pd.DataFrame):
        self.__cycle_station_with_borough_names_preview = value

    @property
    def total_duartion_per_borough(self):
        return self.__total_duartion_per_borough

    @total_duartion_per_borough.setter
    def total_duartion_per_borough(self, value: pd.DataFrame):
        self.__total_duartion_per_borough = value

    @property
    def h2o_leaderboard(self):
        return self.__h2o_leaderboard

    @h2o_leaderboard.setter
    def h2o_leaderboard(self, value):
        self.__h2o_leaderboard = value

    @property
    def h2o_leaderboard_df(self):
        return self.__h2o_leaderboard_df

    @h2o_leaderboard_df.setter
    def h2o_leaderboard_df(self, value):
        self.__h2o_leaderboard_df = value
```
